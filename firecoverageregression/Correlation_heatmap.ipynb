```python
import seaborn as sns
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
fire_df = pd.read_csv('FW_Veg_Rem_Combined.csv')
columns_to_keep = ["latitude", "longitude", "Vegetation", "Temp_pre_7", "Temp_cont",
                   "Wind_pre_7", "Wind_cont", "Hum_pre_7", "Hum_cont", "Prec_pre_7", "Prec_cont", "fire_mag"]

clean_df = fire_df[columns_to_keep]
correlation_matrix = clean_df.corr()
corr_fire_mag = correlation_matrix[['fire_mag']]
abs_corr_fire_mag = np.abs(corr_fire_mag)
abs_corr_fire_mag.columns = ['|r(fire_mag)|']
abs_corr_fire_mag.sort_values(by='|r(fire_mag)|', ascending=False, inplace=True)
```

```python
sns.heatmap(abs_corr_fire_mag)
```
